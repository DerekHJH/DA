```python
import os
import matplotlib.pyplot as plt
import torch
```

```python
dataset_name = 'amazon->webcam'
model_save_path = f'./models/{dataset_name}'
title = f'dataset: {dataset_name}; method: DAViT'
device = torch.device('cpu')
```

```python
if os.path.exists(model_save_path):
    checkpoint = torch.load(model_save_path, map_location = device)
    dfhistory = checkpoint['dfhistory']
```

```python
best_accu_s = dfhistory.loc[len(dfhistory) - 1, "best_accu_s"]
best_accu_t = dfhistory.loc[len(dfhistory) - 1, "best_accu_t"]
print(f"the best source accuracy is {best_accu_s}")
print(f"the best target accuracy is {best_accu_t}")
```

```python
plt.plot(dfhistory["err_s_domain"], color = 'red', label = 'source domain loss')
plt.plot(dfhistory["err_t_domain"], color = 'blue', label = 'target cdomain loss')
plt.plot(dfhistory["err_s_label"], color = 'black', label = 'source label loss')
plt.plot(dfhistory["err"], color = 'green', label = 'total loss: addition of the three')
plt.legend();
plt.xlabel('number of epochs')
plt.ylabel('loss')
plt.title(title)
plt.show()
```

```python
plt.plot(dfhistory["accu_s"], color = 'red', label = 'source accuracy')
plt.plot(dfhistory["accu_t"], color = 'blue', label = 'target accuracy')
plt.legend();
plt.xlabel('number of epochs')
plt.ylabel('accuracy')
plt.title(title)
plt.show()
```
